```python
# Feature Extraction with Univariate Statistical Tests (Chi-squared for classification)
from pandas import read_csv
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2


dfArr = df1.values
X = dfArr[:,1:12]
Y = dfArr[:,12]

# feature extraction
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)

# run a wquick model to get accuracy score
# summarize selected features
print(features[0:10,:])
```

```python
selectedFeatures = pd.DataFrame(features)

selectedFeatures
```

```python
subsetCols = ['citric acid','free sulfur dioxide','total sulfur dioxide', 'alcohol']

selectedFeatures.columns = subsetCols
selectedFeatures
```

```python
# Feature Extraction with RFE
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

dfArr = df1.values
X = dfArr[:,1:12]
Y = dfArr[:,12]
# feature extraction
model = LogisticRegression(max_iter = 5000 )
#model = GaussianNB()
rfe = RFE(model,n_features_to_select=4)
fit = rfe.fit(X, Y)
print("Num Features: {}".format(fit.n_features_)) 
print("Selected Features: {}".format(fit.support_))
print("Feature Ranking: {}".format(fit.ranking_)) 
```

```python
Principal Component Analysis (or PCA) uses linear algebra to transform the dataset into a
compressed form. 

Generally this is called a data reduction technique. 
A property of PCA is that you can choose the number of dimensions or principal components 
in the transformed result. 
In the example below, we use PCA and select 3 principal components.
```

```python
Bagged decision trees like Random Forest and Extra Trees 
can be used to estimate the importance of features. 

In the example below we construct a ExtraTreesClassifier classiffier for the Pima
Indians onset of diabetes dataset. 

 
```

```python
# Feature Importance with Extra Trees Classifier
from pandas import read_csv
from sklearn.ensemble import ExtraTreesClassifier
# load data
'''filename = 'c:/dataset/master/pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(filename, names=names)'''

array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
# feature extraction
model = ExtraTreesClassifier(3)
model.fit(X, Y)
print(model.feature_importances_)
```

```python
You can see that we are given an importance score for each attribute where the larger the
score, the more important the attribute. 
The scores suggest the importance of plas, age and mass.
[ 0.11070069 0.2213717'
```
